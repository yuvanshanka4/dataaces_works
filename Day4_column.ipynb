```python
import pandas as pd
from transformers import BertTokenizer, BertForMaskedLM

# Load the Excel file into a DataFrame
file_path = "Dataset1.xlsx"
df = pd.read_excel(file_path)

# Load the BERT tokenizer and model
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForMaskedLM.from_pretrained(model_name)

# Function to generate summaries for articles
def generate_summary(article_text):
    inputs = tokenizer.encode(article_text, return_tensors="pt", max_length=512, truncation=True)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer.encode(article_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model(inputs, labels=labels)
    loss = outputs.loss
    predicted_tokens = tokenizer.decode(outputs.logits.argmax(dim=-1)[0], skip_special_tokens=True)
    return predicted_tokens

# Create a new column 'summary' and populate it with the generated summaries
df['summary'] = df['content'].apply(generate_summary)

# Print the DataFrame with the new 'summary' column
print(df)
```

```python
import pandas as pd
from transformers import BartTokenizer, BartForConditionalGeneration

# Load the BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

# Function to summarize the text
def summarize_text(text):
    # Tokenize the input text
    inputs = tokenizer(text, max_length=1024, return_tensors='pt', truncation=True)

    # Generate the summary
    summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=30, max_length=150, early_stopping=True)

    # Decode the generated summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Read the Excel file into a DataFrame
excel_file_path = '/content/Dataset1.xlsx'
df = pd.read_excel(excel_file_path)

# Initialize an empty list to store the summaries
summaries = []

# Iterate through each row and generate summaries one by one
for index, row in df.iterrows():
    content = row['content']
    summary = summarize_text(content)
    print(summary)
    summaries.append(summary)

    # Add the summary to the DataFrame for the current row
    df.at[index, 'summary'] = summary

# Save the updated DataFrame back to the Excel file
updated_excel_file_path = '/content/UpdatedDataset1.xlsx'
df.to_excel(updated_excel_file_path, index=False)

print("Summaries added and Excel file updated.")
```

```python
import pandas as pd
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the T5 tokenizer and model
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')

# Function to generate summaries for text
def generate_summary(text):
    inputs = tokenizer.encode("summarize: " + text, return_tensors='pt', max_length=512, truncation=True, add_special_tokens=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=50, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Read the Excel file into a DataFrame
excel_file_path = 'Dataset1.xlsx'
df = pd.read_excel(excel_file_path)

# Initialize an empty list to store the summaries
summaries = []

# Iterate through each row and generate summaries one by one
for index, row in df.iterrows():
    content = row['content']
    summary = generate_summary(content)
    summaries.append(summary)

    # Add the summary to the DataFrame for the current row
    df.at[index, 'summary'] = summary

# Save the updated DataFrame back to the Excel file
updated_excel_file_path = 'UpdatedDataset12.xlsx'
df.to_excel(updated_excel_file_path, index=False)

print("Summaries added and Excel file updated.")
```
