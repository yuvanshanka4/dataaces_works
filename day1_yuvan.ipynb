```python
from PyPDF2 import PdfReader #to import pdf
from langchain.embeddings.openai import OpenAIEmbeddings #to connect the embeddings with the model
from langchain.text_splitter import CharacterTextSplitter # to split the characted with some word limit
from langchain.vectorstores import FAISS #generates a QandA model

import os
os.environ["OPENAI_API_KEY"] = "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" #API key
```

```python
pdf=PdfReader("Module-7 NoSQL.pdf")
```

```python
#reading text from the pdf
from typing_extensions import Concatenate
# read text from pdf
raw = ''
for i, page in enumerate(pdf.pages):
    content = page.extract_text()
    if content:
        raw += content
raw
```

```python
#removing \n
split = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 800,
    chunk_overlap  = 200,
    length_function = len,
)
texts =split.split_text(raw)
len(texts)
```

```python
#finding the embedding vector for the text above
embeddings = OpenAIEmbeddings()
document_search = FAISS.from_texts(texts, embeddings) #matches the text and its embeddings
document_search
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

```python
#query input and it output
chain = load_qa_chain(OpenAI(), chain_type="stuff")
a=1
while a==1:
  query = input("User query:")
  if query.lower()=="stop":
    a=0
    print("ended")
  docs = document_search.similarity_search(query)
  print("answer:"+chain.run(input_documents=docs, question=query))
```
